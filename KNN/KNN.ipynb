```
import pandas as pd
```

```
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```
names = ["sepal-length", "sepal-width", "petal-length", "petal-width", "class"]
```

```
dataset = pd.read_csv(url, names=names)
```

```
dataset
```

```
dataset.head()
```

```
x = dataset.drop("class", axis=1)
```

```
x
```

```
y = dataset["class"]
```

```
y
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2) 
```

```
from sklearn.preprocessing import StandardScaler
sca = StandardScaler()
sca.fit(x_train)
x_train = sca.transform(x_train)
x_test = sca.transform(x_test)
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
clf = KNeighborsClassifier(n_neighbors=5)
```

```
clf.fit(x_train, y_train)
```

```
pred = clf.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix, classification_report
```

```
print(confusion_matrix(y_test, pred))
```

```
print(classification_report(y_test, pred))
```

### Comparing error values with different n values

```
import numpy as np
error = []
for i in range(1, 40):
  k = KNeighborsClassifier(n_neighbors=i)
  k.fit(x_train, y_train)
  pred = k.predict(x_test)
  error.append(np.mean(pred != y_test))
  
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(range(1, 40), error, color='blue', linestyle= 'dashed', marker= 'o', markerfacecolor= 'red', markersize= 10)
plt.title("Error rate for K values")
plt.xlabel("K values")
plt.ylabel("Mean Error")
```
